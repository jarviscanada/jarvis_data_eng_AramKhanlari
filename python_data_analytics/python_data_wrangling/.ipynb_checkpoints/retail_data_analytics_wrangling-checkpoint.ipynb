Visualize total invoice amount (both negative values included and excluded cases).

````python
retail_df['invoice_total'] = retail_df['quantity'] * retail_df['unit_price']
invoice_amounts = retail_df.groupby('invoice_no')['invoice_total'].sum()
invoice_status = invoice_amounts >= 0
print(invoice_amounts.head())
print(invoice_status.head())
````

````python
def show_distribution(var_data):
    from matplotlib import pyplot as plt
    # Get statistics
    min_val = var_data.min()
    max_val = var_data.max()
    mean_val = var_data.mean()
    med_val = var_data.median()
    mod_val = var_data.mode()[0]

    print('Minimum:{:.2f}\nMean:{:.2f}\nMedian:{:.2f}\nMode:{:.2f}\nMaximum:{:.2f}\n'.format(min_val,
                                                                                            mean_val,
                                                                                            med_val,
                                                                                            mod_val,
                                                                                            max_val))

    # Create a figure for 2 subplots (2 rows, 1 column)
    fig, ax = plt.subplots(2, 1, figsize = (10,4))

    # Plot the histogram   
    ax[0].hist(var_data)
    ax[0].set_ylabel('Frequency')

    # Add lines for the mean, median, and mode
    ax[0].axvline(x=min_val, color = 'gray', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=mean_val, color = 'cyan', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=med_val, color = 'red', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=mod_val, color = 'yellow', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=max_val, color = 'gray', linestyle='dashed', linewidth = 2)

    # Plot the boxplot   
    ax[1].boxplot(var_data, vert=False)
    ax[1].set_xlabel('Value')

    # Add a title to the Figure
    fig.suptitle('Data Distribution')

    # Show the figure
    fig.show()

# Get the variable to examine
col = retail_df[retail_df.invoice_total>0]['invoice_total']
# Call the function
show_distribution(col)
# Get the variable to examine
col = retail_df['invoice_total']
# Call the function
show_distribution(col)
````

````python
q85 = invoice_amounts.quantile(0.85)
filtered_invoice_amounts = invoice_amounts[invoice_amounts>0][invoice_amounts <= q85]
show_distribution(filtered_invoice_amounts)
filtered_invoice_amounts = invoice_amounts[invoice_amounts <= q85]
show_distribution(filtered_invoice_amounts)
````

````python
retail_df['yyyymm'] = retail_df['invoice_date'].dt.year * 100 + retail_df['invoice_date'].dt.month
````

````python
invoice_amounts = retail_df.groupby(['invoice_no', 'yyyymm'])['invoice_total'].sum()
retail_df['invoice_status'] = retail_df['invoice_total'] >= 0
````

````python
monthly_canceled = (
    retail_df[retail_df['invoice_status'] == False]  
        .groupby('yyyymm')['invoice_no']
        .nunique()                                    
        .rename("canceled_orders")
)
#print(monthly_canceled.head())
````

````python
monthly_total = (
    invoice_amounts.groupby('yyyymm')
    .size()
    .rename("total_invoices")
)
# print(monthly_total.head())
````

````python
monthly_orders = pd.concat([monthly_total, monthly_canceled], axis=1)
monthly_orders.plot( kind='bar', figsize=(15,10))
````

````python
first_purchase = (
    retail_df.groupby('customer_id')['yyyymm']
    .min()
    .rename('first_yyyymm')
)
````

````python
retail_df['is_new'] = retail_df['yyyymm'] == retail_df['customer_id'].map(first_purchase)
````

````python
monthly_new = (
    retail_df[retail_df['is_new']]
        .groupby('yyyymm')['customer_id']
        .nunique()
        .rename("new_users")
)
monthly_existing = (
    retail_df[~retail_df['is_new']]
        .groupby('yyyymm')['customer_id']
        .nunique()
        .rename("existing_users")
)
monthly_users = pd.concat([monthly_new, monthly_existing], axis=1)
````

````python
monthly_users.plot(
    kind='bar',
    figsize=(15,8),
    stacked=False
)
plt.title("New vs Existing Users per Month")
plt.xlabel("Year-Month")
plt.ylabel("Number of Users")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
````

## Finding RFM

RFM is a method used for analyzing customer value. It is commonly used in database marketing and direct marketing and has received particular attention in the retail and professional services industries. ([wikipedia](https://en.wikipedia.org/wiki/RFM_(market_research)))

Optional Reading: [Making Your Database Pay Off Using Recency Frequency and Monetary Analysis](http://www.dbmarketing.com/2010/03/making-your-database-pay-off-using-recency-frequency-and-monetary-analysis/)


RFM stands for three dimensions:

- Recency  How recently did the customer purchase?

- Frequency  How often do they purchase?

- Monetary Value  How much do they spend?

<!-- Note: To simplify the problem, let's keep all placed and canceled orders.


**Sample RFM table**

![](https://i.imgur.com/sXFIg6u.jpg) -->

````python
df_recency = retail_df.groupby(by='customer_id', as_index=False)['invoice_date'].max()
df_recency.columns = ['customer_id', 'LastPurchaseDate']
recent_date = df_recency['LastPurchaseDate'].max()
df_recency['Recency'] = df_recency['LastPurchaseDate'].apply(lambda x: (recent_date - x).days)
df_recency.head()
````

````python
frequency_df = retail_df.drop_duplicates().groupby(by=['customer_id'], as_index=False)['invoice_date'].count()
frequency_df.columns = ['customer_id', 'Frequency']
frequency_df.head()
````

````python
retail_df['Total'] = retail_df['invoice_total']  # Total spent by each customer
monetary_df = retail_df.groupby(by='customer_id', as_index=False)['Total'].sum()
monetary_df.columns = ['customer_id', 'Monetary']
monetary_df.head()
````

````python
rf_df = df_recency.merge(frequency_df, on='customer_id')
rfm_table= rf_df.merge(monetary_df, on='customer_id').drop(columns='LastPurchaseDate')
rfm_table.head()
````

# RFM Segmentation

---
**Please remove this insturction cell after you are done with coding**
RFM segmentation categorizes your customers into different segments, according to their interactions with your website, which will allow you to subsequently approach these groups in the most effective way. In this article, we will show you how to make an RFM segmentation based on an RFM score combining all three RFM parameters together and allowing you to divide your customers into 11 different segments. 

- [RFM Segmentation business cases](https://docs.exponea.com/docs/rfm-segmentation-business-use)

- [RFM Segmentation Guide](https://docs.exponea.com/docs/rfm-segmentation-business-use)

As you can see, computing RFM segmentation requires extensive domain knowledge in marketing which is out of the scope in this project.
````python
#RFM score values 
rfm_table['RecencyScore'] = pd.qcut(rfm_table['Recency'],5,labels=[5,4,3,2,1])
rfm_table['FrequencyScore'] = pd.qcut(rfm_table['Frequency'].rank(method="first"),5,labels=[1,2,3,4,5])
rfm_table['MonetaryScore'] = pd.qcut(rfm_table['Monetary'],5,labels=[1,2,3,4,5])
````

````python
rfm_table.head()
````

````python
#RFM score values are combined side by side in str format
(rfm_table['RecencyScore'].astype(str) + 
 rfm_table['FrequencyScore'].astype(str) + 
 rfm_table['MonetaryScore'].astype(str)).head()
````

````python
#calculation of the RFM score
rfm_table["RFM_SCORE"] = rfm_table['RecencyScore'].astype(str) + rfm_table['FrequencyScore'].astype(str) + rfm_table['MonetaryScore'].astype(str)
````

````python
rfm_table.describe().T
````

````python
#customers with RFM Score 555
rfm_table[rfm_table["RFM_SCORE"] == "555"].head()
````

````python
#customers with RFM Score 111
rfm_table[rfm_table["RFM_SCORE"] == "111"].head()
````

````python
#segmenting of customers according to RecencyScore and FrequencyScore values
seg_map = {
    r'[1-2][1-2]': 'Hibernating',
    r'[1-2][3-4]': 'At Risk',
    r'[1-2]5': 'Can\'t Lose',
    r'3[1-2]': 'About to Sleep',
    r'33': 'Need Attention',
    r'[3-4][4-5]': 'Loyal Customers',
    r'41': 'Promising',
    r'51': 'New Customers',
    r'[4-5][2-3]': 'Potential Loyalists',
    r'5[4-5]': 'Champions'
}
````

````python
#something extra 
rfm_labels = [
    ["Can't Lose", "Can't Lose", "Loyal Customers", "Loyal Customers", "Champions"],  # F = 1
    ["At Risk", "At Risk", "Loyal Customers", "Loyal Customers", "Champions"],        # F = 2
    ["At Risk", "At Risk", "Need Attention", "Potential Loyalists", "Potential Loyalists"],  # F = 3
    ["Hibernating", "Hibernating", "About to Sleep", "Potential Loyalists", "Potential Loyalists"],    # F = 4
    ["Hibernating", "Hibernating", "About to Sleep", "Promising", "New Customers"] # F = 5 (51 bottom-right)
]

# Convert labels to numeric IDs
unique_segments = list({seg for row in rfm_labels for seg in row})
segment_to_id = {seg: i for i, seg in enumerate(unique_segments)}
rfm_numeric = np.array([[segment_to_id[seg] for seg in row] for row in rfm_labels])

plt.figure(figsize=(10, 8))
plt.imshow(rfm_numeric, cmap="viridis")
plt.colorbar(label="Segment Code")

for i in range(5):
    for j in range(5):
        plt.text(j, i, rfm_labels[i][j], ha='center', va='center', fontsize=8, color="white")

plt.title("RFM Heatmap (Recency 1?5 left?right, Frequency 1?5 top?bottom)")
plt.xlabel("Recency Score (1 ? 5)")
plt.ylabel("Frequency Score (1 ? 5)")

plt.xticks(np.arange(5), ["1", "2", "3", "4", "5"])
plt.yticks(np.arange(5), ["1", "2", "3", "4", "5"])

plt.tight_layout()
plt.show()
````

````python
#creation of segment variable
rfm_table['Segment'] = rfm_table['RecencyScore'].astype(str) + rfm_table['FrequencyScore'].astype(str)
rfm_table['Segment'] = rfm_table['Segment'].replace(seg_map, regex=True)
````
